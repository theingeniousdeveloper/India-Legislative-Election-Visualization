```python
import pandas as pd
```

```python
votes = pd.read_csv("Legislative.csv" )
votes.head(3)
```

```python
votes.info()
```

```python
votes = votes.apply(lambda x: x.fillna(0) if x.dtype.kind in 'biufc' else x.fillna('NA'))
```

```python
votes.head(3)
```

```python
votes.nunique()
```

```python
votes[["STATE","GENDER","WINNER","CATEGORY"]]=votes[["STATE","GENDER","WINNER","CATEGORY"]].astype("category")
```

```python
votes.rename({"OVER TOTAL ELECTORS \nIN CONSTITUENCY":"PERCENTAGE OF TOTAL ELECTORS VOTED FOR THE CANDIDATE ","OVER TOTAL VOTES POLLED \nIN CONSTITUENCY":"PERCENTAGE OF TOTAL VOTES RECEIVED BY THE CANDIDATE"},axis="columns",inplace=True)
```

```python
votes=votes.round(1)
```

```python
votes[["CONSTITUENCY","NAME"]] = votes[["CONSTITUENCY","NAME"]].apply(lambda x: x.str.title())
```

```python
votes.head()
```

```python
votes["ASSETS"] = votes["ASSETS"].str.split("\n").str[0].str.split(" ").str[1]
votes["LIABILITIES"] = votes["LIABILITIES"].str.split("\n").str[0].str.split(" ").str[1]
```

```python
votes.to_csv("file2.csv")
```
